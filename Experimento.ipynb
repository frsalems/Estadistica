### Uso de Ruidos para la generación de Modelos de Mínimos Cuadrados Ordinarios

Primeramente se importaron todos los paquetes que se van a utilizar.

```
import statsmodels.api as sm
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import statistics as stats
from numpy.random import default_rng

rng = default_rng(12345)
```

En la **línea 8** de la celda de arriba la variable **rng** tiene un método para obtener una nueva instancia de un generador de números, los números dentro simbolizan el parámetro de la semilla de inicialización.

En la celda de abajo, se comienza por generar **30 números** entre los valores de **-4** y **4** para **X**, seguido de esto se mezclan para que tengan un comportamiento más aleatorio.

A partir de estos datos se obtuvieron las medidas de tendencia central: **Media** (*Mean*), **Mediana** (*Median*) y **Desv. Estandar** (*Desviation Standard - Scale*) las cuales serán utilizadas para obtener datos más adelante. En conjunto con las medidas de tendencia central, se obtuvo la **Tendencia** y la **Traslación** pero para ambos valores fue utilizada una ecuación de un libro que sugirió el maestro, por lo tanto, los valores descritos corresponden a los del libro.

Seguido de esto, se determinó la ecuación de **Y** en su respectiva variable en donde se utilizó los valores generados para **X** así como los valores de la **Tendencia** y la **Traslación**.

Finalmente, se crearon **8** diferentes ruidos por cada tipo de ruido, siendo estos: **Uniforme**, **Poisson**, **Laplace** y **Normal**, en cada uno se aplicaron diferentes valores para tener ruidos más pronunciados y así obtener diferentes resultados.

```
# Data
x = np.linspace(-4, 4, 30)
rng.shuffle(x)

# Mean
lam = x.mean()
# Median
loc = stats.median(x)
# Desviation Standard
scale = stats.pstdev(x)

# Tendencia
#trend = 1.5
trend = 4.3599
# Traslación
#shift = 2
shift = 3.5249

# Y's values
y = trend*x + shift

# Noises
# ------ Uniform
u1 = rng.uniform(-2.0, 2.0, size=30)
u2 = rng.uniform(-1.6, 1.6, size=30)
u3 = rng.uniform(-1.2, 1.2, size=30)
u4 = rng.uniform(-0.9, 0.9, size=30)
u5 = rng.uniform(-0.6, 0.6, size=30)
u6 = rng.uniform(-0.4, 0.4, size=30)
u7 = rng.uniform(-0.2, 0.2, size=30)
u8 = rng.uniform(-0.1, 0.1, size=30)
# ------ Poisson
p1 = rng.poisson(lam=0, size=30)
p2 = rng.poisson(lam=0.1, size=30)
p3 = rng.poisson(lam=0.3, size=30)
p4 = rng.poisson(lam=0.6, size=30)
p5 = rng.poisson(lam=1.0, size=30)
p6 = rng.poisson(lam=1.5, size=30)
p7 = rng.poisson(lam=2.1, size=30)
p8 = rng.poisson(lam=2.8, size=30)
# ------ Laplace
l1 = rng.laplace(loc=0, scale=1, size=30)
l2 = rng.laplace(loc=0.2, scale=1, size=30)
l3 = rng.laplace(loc=0.5, scale=1, size=30)
l4 = rng.laplace(loc=0.9, scale=1, size=30)
l5 = rng.laplace(loc=1.4, scale=1, size=30)
l6 = rng.laplace(loc=2.0, scale=1, size=30)
l7 = rng.laplace(loc=2.7, scale=1, size=30)
l8 = rng.laplace(loc=3.5, scale=1, size=30)
# ------ Normal
n1 = rng.normal(loc=0, scale=1, size=30)
n2 = rng.normal(loc=0.2, scale=1, size=30)
n3 = rng.normal(loc=0.5, scale=1, size=30)
n4 = rng.normal(loc=0.9, scale=1, size=30)
n5 = rng.normal(loc=1.4, scale=1, size=30)
n6 = rng.normal(loc=2.0, scale=1, size=30)
n7 = rng.normal(loc=2.7, scale=1, size=30)
n8 = rng.normal(loc=3.5, scale=1, size=30)
```

En esta parte se obtuvieron las $R^2$ a partir de una función, primeramente se habían calculado de esta forma pero después se optó por hacerlo con un modelo de **Mínimos Cuadrados Ordinarios** (*OLS*).

### $R^2$ solo con función

```
from sklearn.metrics import r2_score
```

```
# R^2 Scores
# ----- Uniform
ru1 = r2_score(y, y+u1)
ru2 = r2_score(y, y+u2)
ru3 = r2_score(y, y+u3)
ru4 = r2_score(y, y+u4)
ru5 = r2_score(y, y+u5)
ru6 = r2_score(y, y+u6)
ru7 = r2_score(y, y+u7)
ru8 = r2_score(y, y+u8)
# ----- Poisson
rp1 = r2_score(y, y+p1)
rp2 = r2_score(y, y+p2)
rp3 = r2_score(y, y+p3)
rp4 = r2_score(y, y+p4)
rp5 = r2_score(y, y+p5)
rp6 = r2_score(y, y+p6)
rp7 = r2_score(y, y+p7)
rp8 = r2_score(y, y+p8)
# ----- Laplace
rl1 = r2_score(y, y+l1)
rl2 = r2_score(y, y+l2)
rl3 = r2_score(y, y+l3)
rl4 = r2_score(y, y+l4)
rl5 = r2_score(y, y+l5)
rl6 = r2_score(y, y+l6)
rl7 = r2_score(y, y+l7)
rl8 = r2_score(y, y+l8)
# ----- Normal
rn1 = r2_score(y, y+n1)
rn2 = r2_score(y, y+n2)
rn3 = r2_score(y, y+n3)
rn4 = r2_score(y, y+n4)
rn5 = r2_score(y, y+n5)
rn6 = r2_score(y, y+n6)
rn7 = r2_score(y, y+n7)
rn8 = r2_score(y, y+n8)
```

Para este punto se calcularon las $R^2$ con una función propia del Modelo de Mínimos Cuadrados, con el objetivo de obtener $R^2$ más realistas y más cercanas a cada tipo de variación en los ruidos y en sus tipos.

Por ello, primero se creó la variable de **pred_x** donde se le añadió a la variable **X** una columna de 1 para que tuviera un valor constante en todos los números, si bien puede sonar extraño es parte del proceso para generar el modelo de **OLS**.

La tabla mostrada al final es para que se aprecie que viene incluído el valor de $R^2$.

### $R^2$ con OLS

```
import statsmodels.api as sm

pred_x = sm.add_constant(x)
```

```
modelu1 = sm.OLS(y+u1, pred_x).fit()
modelu2 = sm.OLS(y+u2, pred_x).fit()
modelu3 = sm.OLS(y+u3, pred_x).fit()
modelu4 = sm.OLS(y+u4, pred_x).fit()
modelu5 = sm.OLS(y+u5, pred_x).fit()
modelu6 = sm.OLS(y+u6, pred_x).fit()
modelu7 = sm.OLS(y+u7, pred_x).fit()
modelu8 = sm.OLS(y+u8, pred_x).fit()

modelp1 = sm.OLS(y+p1, pred_x).fit()
modelp2 = sm.OLS(y+p2, pred_x).fit()
modelp3 = sm.OLS(y+p3, pred_x).fit()
modelp4 = sm.OLS(y+p4, pred_x).fit()
modelp5 = sm.OLS(y+p5, pred_x).fit()
modelp6 = sm.OLS(y+p6, pred_x).fit()
modelp7 = sm.OLS(y+p7, pred_x).fit()
modelp8 = sm.OLS(y+p8, pred_x).fit()

modell1 = sm.OLS(y+l1, pred_x).fit()
modell2 = sm.OLS(y+l2, pred_x).fit()
modell3 = sm.OLS(y+l3, pred_x).fit()
modell4 = sm.OLS(y+l4, pred_x).fit()
modell5 = sm.OLS(y+l5, pred_x).fit()
modell6 = sm.OLS(y+l6, pred_x).fit()
modell7 = sm.OLS(y+l7, pred_x).fit()
modell8 = sm.OLS(y+l8, pred_x).fit()

modeln1 = sm.OLS(y+n1, pred_x).fit()
modeln2 = sm.OLS(y+n2, pred_x).fit()
modeln3 = sm.OLS(y+n3, pred_x).fit()
modeln4 = sm.OLS(y+n4, pred_x).fit()
modeln5 = sm.OLS(y+n5, pred_x).fit()
modeln6 = sm.OLS(y+n6, pred_x).fit()
modeln7 = sm.OLS(y+n7, pred_x).fit()
modeln8 = sm.OLS(y+n8, pred_x).fit()

print(modell3.summary())
#print(modelu1.rsquared)
```

Para poder generar los valores que serán la predicción en los modelos, primero se creó la variable de **model_x1** en donde se realiza una acción similar a la anterior, pero ahora con 50 valores entre -4 y 4 para obtener un modelo más preciso.

#### Creación de modelo de X

```
model_x1 = sm.add_constant(np.linspace(-4, 4))
```

Con los valores del **model_x1** se calculan todas las Y's correspondientes para las variaciones de los ruidos de **Laplace** y **Normal**.

#### Creación de los modelos de Y

```
model_yl1 = modell1.predict(model_x1)
model_yl2 = modell2.predict(model_x1)
model_yl3 = modell3.predict(model_x1)
model_yl4 = modell4.predict(model_x1)
model_yl5 = modell5.predict(model_x1)
model_yl6 = modell6.predict(model_x1)
model_yl7 = modell7.predict(model_x1)
model_yl8 = modell8.predict(model_x1)

model_yn1 = modeln1.predict(model_x1)
model_yn2 = modeln2.predict(model_x1)
model_yn3 = modeln3.predict(model_x1)
model_yn4 = modeln4.predict(model_x1)
model_yn5 = modeln5.predict(model_x1)
model_yn6 = modeln6.predict(model_x1)
model_yn7 = modeln7.predict(model_x1)
model_yn8 = modeln8.predict(model_x1)
```
